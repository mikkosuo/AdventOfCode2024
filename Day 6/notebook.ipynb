```python
lab = []
with open('input.txt') as file:
    for line in file:
        lab.append([c for c in line.strip()])
```

```python
StartLocation = [-1, -1]
for i in range(len(lab)):
    for j in range(len(lab[i])):
        if(lab[i][j] == "^"):
            StartLocation = [i, j]
print(StartLocation)
```

```python
def new_direction(Old):
    dirs = ["up", "right", "down", "left"]
    next_index = (dirs.index(Old) + 1) % 4
    return dirs[next_index] 

def step(i, j, direction):
    out_of_map = False
    next_state = "unknown"
    while next_state != ".":
        iPropose = i
        jPropose = j
        if direction == "up":
            if i == 0:
                out_of_map = True
                break
            iPropose-=1
        if direction == "down":
            if i == len(lab)-1:
                out_of_map = True
                break
            iPropose+=1
        if direction == "left":
            if j == 0:
                out_of_map = True
                break
            jPropose-=1
        if direction == "right":
            if j == len(lab[i])-1:
                out_of_map = True
                break
            jPropose+=1
        next_state = lab[iPropose][jPropose]
        if(next_state == "#"):
            new_dir = new_direction(direction)
            direction = new_dir
        else:
            i = iPropose
            j = jPropose
    return i, j, direction, out_of_map    

direction = "up"
distinct_visited_spots = []
i = StartLocation[0]
j = StartLocation[1]
while True:
    distinct_visited_spots.append((i,j))
    i, j, direction, out_of_map = step(i, j, direction)
    if out_of_map:
        break
    
print(len(set(distinct_visited_spots)))
print(distinct_visited_spots)
```

```python
# Part 2:
num_of_valid_places = 0
for (i, j) in set(distinct_visited_spots):
    print(i,j)
    if [i, j] == StartLocation:
        continue
    if lab[i][j] == "#":
        continue
    # Change the map temporarily
    lab[i][j] = "#"
    # Step until out of map or repeateted positions
    loop_positions = []
    iLoop = StartLocation[0]
    jLoop = StartLocation[1]
    directionLoop = "up"
    out_of_map = False
    while True:
        loop_positions.append(((iLoop,jLoop), directionLoop))
        iLoop, jLoop, directionLoop, out_of_map = step(iLoop, jLoop, directionLoop)
        if out_of_map:
            break
        if ((iLoop,jLoop), directionLoop) in loop_positions:
            num_of_valid_places += 1
            break
    # Revert the temporary map change
    lab[i][j] = "."
print(num_of_valid_places)
```
