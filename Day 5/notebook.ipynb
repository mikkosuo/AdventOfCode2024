```python
restrictions = {}
rules = []
with open('input.txt') as file:
    for line in file:
        lineAsList = line.strip().split('|')
        if(len(lineAsList) == 2):
            # Reading restrictions
            first = int(lineAsList[0])
            second = int(lineAsList[1])
            if second in restrictions.keys():
                restrictions[second].append(first)
            else:
                restrictions[second] = [first]
        else:
            # Reading ordering rules
            rules.append([int(i) for i in line.strip().split(",") if i.isdigit()])
```

```python
ValidRules = []
InvalidRules = []
for rule in rules:
    RuleValid = True
    for firstPageNumIndex in range(len(rule)):
        for secondPageNumIndex in range(firstPageNumIndex, len(rule)):
            first = rule[firstPageNumIndex]
            second = rule[secondPageNumIndex]
            if second in restrictions[first]:
                RuleValid = False
    if RuleValid:
        ValidRules.append(rule)
    else:
        InvalidRules.append(rule)
```

```python
sumOfMiddlePageNumbers = 0
for rule in ValidRules:
    if len(rule) < 1:
        continue
    sumOfMiddlePageNumbers += rule[int((len(rule)-1)/2)]
print(sumOfMiddlePageNumbers)
```

```python
FixedRules = []
for rule in InvalidRules.copy():
    fixedRule = []
    rulesLeft = rule.copy()
    while len(rulesLeft) > 0:
        for i in range(len(rulesLeft)):
            if set(restrictions[rulesLeft[i]]).intersection(rulesLeft):
                continue
            else:
                fixedRule.append(rulesLeft[i])
                del rulesLeft[i]
                break
    FixedRules.append(fixedRule)
```

```python
sumOfMiddlePageNumbers = 0
for rule in FixedRules:
    if len(rule) < 1:
        continue
    sumOfMiddlePageNumbers += rule[int((len(rule)-1)/2)]
print(sumOfMiddlePageNumbers)
```
