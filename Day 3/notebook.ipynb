```python
inputString = ""
with open('input.txt') as file:
    inputString = file.read().replace('\n', '')
```

```python
# Part 1
def process_text_after_mul_keyword(InputString, LPosition, Total):
    if LPosition > len(InputString) - 3:
        return
    iterPosition = 1
    iterator = InputString[LPosition + iterPosition]
    if iterator != '(':
        return
    iterPosition += 1
    firstMultiplicant = []
    while True:
        iterator = InputString[LPosition+iterPosition]
        if iterator == ',':
            iterPosition += 1
            break

        if iterator.isdigit():
            firstMultiplicant.append(int(iterator))
            iterPosition += 1
        else:
            return
    secondMultiplicant = []
    while True:
        iterator = InputString[LPosition+iterPosition]
        if iterator == ')':
            break

        if iterator.isdigit():
            secondMultiplicant.append(int(iterator))
            iterPosition += 1
        else:
            return
    
    if len(firstMultiplicant) == 0 or len(secondMultiplicant) == 0:
        return

    number1 = 0

    firstMultiplicant.reverse()
    for i in range(len(firstMultiplicant)):
        number1 += firstMultiplicant[i] * 10**i

    number2 = 0
    secondMultiplicant.reverse()
    for i in range(len(secondMultiplicant)):
        number2 += secondMultiplicant[i] * 10**i
    
    Total[0] += number1 * number2


Total = [0]
for i in range(len(inputString)-3):
    M = inputString[i]
    U = inputString[i+1]
    L = inputString[i+2]
    if M == 'm' and U == 'u' and L == 'l':
        process_text_after_mul_keyword(inputString, i+2, Total)

print(f"Total = {Total[0]}")
```
